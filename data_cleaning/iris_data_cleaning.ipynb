```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../data/iris.csv')
print(df.shape)
df.head()
```

```python
df = df.drop('Id',axis=1)
```

```python
df_pca = df.drop(['Species'],axis=1)
df_pca.head()
```

```python
pca = PCA(n_components = 2)
pca_comps = pd.DataFrame(pca.fit_transform(df_pca))
print(pca.explained_variance_ratio_)
```

```python
df['pc1'] = pca_comps[0]
df['pc2'] = pca_comps[1]
df
```

```python
df.to_csv('../data/iris_pca.csv',index=False)
```
