```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../data/pokemon.csv')
print(df.shape)
df.head()
```

```python
df.columns
```

```python
df.type1.unique()
```

```python
cols = ['name','type1'] 
cols.extend(['against_water','against_grass','against_fire','against_electric','against_rock'])
# cols.extend([x for x in df.columns if 'against' in x])
df_small = df[cols]
print(df_small.shape)
df_small.head()
```

```python
df_less = df_small[df_small['type1'].isin(['water','grass','fire','electric','rock'])]
print(df_less.shape)
df_less.reset_index(inplace=True,drop=True)
df_less.head()
```

```python
df_pca = df_less.drop(['name','type1'],axis=1)
df_pca.head()
```

```python
df_pca.min()
```

```python
df_pca.max()
```

```python
pca = PCA(n_components = 2)
pca_comps = pd.DataFrame(pca.fit_transform(df_pca))
print(pca.explained_variance_ratio_)
```

```python
df_less['pc1'] = pca_comps[0]
df_less['pc2'] = pca_comps[1]
df_less
```

```python
df_less.to_csv('../data/pokemon_small.csv',index=False)
```
