```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../data/pokemon.csv')
print(df.shape)
df.head()
```

```python
df.columns
```

```python
df.type1.unique()
```

```python
cols = ['name','type1'] 
# cols.extend(['against_water','against_grass','against_fire'])
cols.extend([x for x in df.columns if 'against' in x])
df = df[cols]
print(df.shape)
df.head()
```

```python
df_small = df[df['type1'].isin(['water','grass','fire'])]
print(df_small.shape)
df_small
```

```python
df_pca = df_small.drop(['name','type1'],axis=1)
df_pca.head()
```

```python
pca = PCA()
pca.fit(df_pca)
print(pca.explained_variance_ratio_)
```
