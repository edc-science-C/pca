```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/foods.csv')
print(df.shape)
df.head()
```

```python
df['Food'] = df['Food and Serving'].apply(lambda x: x.split(',')[0])
df['Type'] = df['Food Type'].apply(lambda x: x.split(',')[0])
df['Type'] = df['Type'].apply(lambda x: x.split()[0])
df.head()
```

```python
df.columns
```

```python
df = df.drop(['Food and Serving','Food Type'],axis=1)
df = df.drop(['Total Carbohydrate (g)','Saturated Fat'],axis=1)
df.head()
```

```python
df_pca = df.drop(['Type','Food'],axis=1)
# df_pca = df.drop(['pc1','pc2'],axis=1)
df_pca.head()
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df_pca)
```

```python
pca = PCA(n_components = 2)
pca_comps = pd.DataFrame(pca.fit_transform(scaled))
print(pca.explained_variance_ratio_)
```

```python
df['pc1'] = pca_comps[0]
df['pc2'] = pca_comps[1]
df.head()
```

```python
cmap = {'Vegetables':'green','Fruits':'red','Seafood':'blue'}
```

```python
pca_comps['Type'] = df['Type']

fig, ax = plt.subplots()
grouped = pca_comps.groupby('Type')
for key, group in grouped:
    group.plot.scatter(ax=ax, x=0, y=1, label=key, color=cmap[key])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
pca_importances = pd.DataFrame(pca.components_,columns=df_pca.columns)
plt.figure(figsize=(12,6))
sns.heatmap(pca_importances, center=0,cmap="coolwarm",annot=True)
```
