```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/foods.csv')
print(df.shape)
df.head()
```

```python
df['Food'] = df['Food and Serving'].apply(lambda x: x.split(',')[0])
df['Type'] = df['Food Type'].apply(lambda x: x.split(',')[0])
df['Type'] = df['Type'].apply(lambda x: x.split()[0])
df.rename(columns={"Saturated Fat": "Saturated Fat (g)"},inplace=True)
df.head()
```

```python
df = df.drop(['Food and Serving','Food Type'],axis=1)
df.head()
```

```python
df_pca = df.drop(['Type','Food'],axis=1)
# df_pca = df.drop(['pc1','pc2'],axis=1)
df_pca.head()
```

```python
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_pca))
df_scaled.columns = df_pca.columns
```

```python
pca = PCA(n_components = 2)
pca_comps = pd.DataFrame(pca.fit_transform(df_scaled))
print(pca.explained_variance_ratio_)
```

```python
df_scaled['pc1'] = pca_comps[0]
df_scaled['pc2'] = pca_comps[1]
df_scaled['Food'] = df['Food']
df_scaled['Type'] = df['Type']
df_scaled.head()
```

```python
df_scaled.to_csv('../data/foods_clean.csv',index=False)
```

```python
df_all = df_scaled.join(df,rsuffix='_orig')
df_all = df_all.drop(['Food_orig','Type_orig'],axis=1)
df_all
```

```python
df_all.to_csv('../data/foods_clean_all.csv',index=False)
```

```python
cmap = {'Vegetables':'green','Fruits':'red','Seafood':'blue'}
```

```python
pca_comps['Type'] = df['Type']

fig, ax = plt.subplots()
grouped = pca_comps.groupby('Type')
for key, group in grouped:
    group.plot.scatter(ax=ax, x=0, y=1, label=key, color=cmap[key])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
pca_importances = pd.DataFrame(pca.components_,columns=df_pca.columns)
plt.figure(figsize=(18,4))
sns.heatmap(pca_importances, center=0,cmap="coolwarm",annot=True)
```
